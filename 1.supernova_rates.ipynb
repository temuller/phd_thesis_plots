# **Plots for Thesis Introduction**

#### In this notebook I plot the photometry and spectra of different types of supernovae for the introduction of my thesis.
#### I have downloaded *json* files with the SNe data from the open supernova catalog.

```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

import warnings  # ignore some annoying warnings in some parts

#sns.set(font_scale=1.5, context='paper', style='white')
%config InlineBackend.figure_format = 'retina'
plt.rcParams["font.family"] = "Gentium"
plt.rcParams['mathtext.fontset'] = "cm"
```

# Rates

```python
# observed rates
## all
percents1 = [79, 17, 4]
labels1 = 'Ia','II','Ibc'

## SNe Ibc
percents2 = [52, 32, 15]
labels2 = 'Ic','Ib','Ibc-pec'

## SNe II
percents3 = [30, 29, 25, 17]
labels3 = 'II-P','IIn','II-L', 'IIb'

# intrinsic rates
percents4 = [57, 24, 19]
labels4 = 'II','Ia','Ibc'

percents5 = [54, 25, 21]
labels5 = 'Ic','Ibc-pec','Ib'

percents6 = [70, 12, 10, 9]
labels6 = 'II-P','IIb','II-L', 'IIn'

textprops={'fontsize':32, 'color':'k'}
#colors = ['darkmagenta', 'darkcyan', 'gold', 'coral']
colors = ['darkred', 'brown', 'indianred', 'lightcoral']

# plot
fig, ax = plt.subplots(2, 3, figsize=(16, 16))

ax[0, 0].pie(percents1, labels=labels1, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.2)
ax[0, 1].pie(percents2, labels=labels2, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.25, startangle=280)
ax[0, 2].pie(percents3, labels=labels3, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.2)
ax[1, 0].pie(percents4, labels=labels4, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.2)
ax[1, 1].pie(percents5, labels=labels5, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.2, startangle=10)
ax[1, 2].pie(percents6, labels=labels6, autopct='%1.0f%%', textprops=textprops, colors=colors, radius=10, labeldistance=1.2)

fontsize = 40
ax[0, 0].set_title(r'All', fontsize=fontsize)
ax[0, 1].set_title(r'SNe Ibc', fontsize=fontsize)
ax[0, 2].set_title(r'SNe II', fontsize=fontsize)
#ax[1, 1].set_title(r'H-rich SNe $intrinsic$ rate by sub-type', fontsize=fontsize)

ax[0, 0].axis('equal')
ax[0, 1].axis('equal')
ax[0, 2].axis('equal')
ax[1, 0].axis('equal')
ax[1, 1].axis('equal')
ax[1, 2].axis('equal')

plt.subplots_adjust(wspace=-10.0, hspace=None)
#plt.savefig('plots/supernova_rates.pdf', bbox_inches='tight')
plt.show()
```
