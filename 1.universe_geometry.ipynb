```python
import numpy as np
import pandas as pd
import glob

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import cm
import seaborn as sns

from astropy.cosmology import FlatLambdaCDM

#sns.set(font_scale=1.5, context='notebook', style='white')
%config InlineBackend.figure_format = 'retina'
plt.rcParams["font.family"] = "Gentium"
plt.rcParams['mathtext.fontset'] = "cm"
```

```python
cmap = cm.summer

fig = plt.figure(num=1, clear=True, figsize=(10, 8), frameon=False)

# flat surface
(x, y) = np.meshgrid(np.arange(-2, 2.1, .1), np.arange(-1, 1.1, .25))
z = x + y

ax = fig.add_subplot(233, projection='3d')
ax.plot_surface(x, y, z, cmap=cmap, vmin=0, vmax=3)
ax.view_init(30, 160)
# Hide grid lines
ax.grid(False)
# Hide axes ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.axis("off")

# sphere
u = np.linspace(0, np.pi, 30)
v = np.linspace(0, 5 * np.pi, 30)

x = np.outer(np.sin(u), np.sin(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.cos(u), np.ones_like(v))

ax = fig.add_subplot(231, projection='3d')
ax.plot_surface(x, y, z, cmap=cmap, vmin=0.0, vmax=None)
ax.view_init(30, 180)
# Hide grid lines
ax.grid(False)
# Hide axes ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.axis("off")

# hyperbola
(x, y) = np.meshgrid(np.arange(-2, 2.1, .1), np.arange(-1, 1.1, .25))
z = x**2 - y**2

ax = fig.add_subplot(232, projection='3d')
ax.plot_surface(x, y, z, cmap=cmap, vmin=0, vmax=5)
ax.view_init(30, 40)
# Hide grid lines
ax.grid(False)
# Hide axes ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.axis("off")

fig.tight_layout()
plt.savefig('plots/universe_geometry.pdf', bbox_inches='tight')
plt.show()
```
